```python
geo_df = mapdata.copy()

# 지도를 초기화 해줄 때 어디를 중심으로 보여줄지 설정합니다.
# 우리가 가져온 데이터프레임 안에 있는 데이터를 기준으로 출력할 수 있도록 위경도의 평균값을 구해옵니다.
map = folium.Map(location=[geo_df['latitude'].median(), geo_df['longitude'].median()], zoom_start=12)

for n in geo_df.index:
    folium.CircleMarker([geo_df['latitude'][n], geo_df['longitude'][n]], 
                  radius=2, icon=folium.Icon(color="yellow")).add_to(map)

map
```
